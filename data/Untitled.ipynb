```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("data.csv")
df
```

```python
numeric_columns = df.select_dtypes(include=[int, float]).columns
filtered_df = df[df[numeric_columns] >= 50]
filtered_df
```

```python
x = list(range(len(df['GO: (대한민국)'])))
x1 = list(range(len(df['python: (대한민국)'])))
x2 = list(range(len(df['C: (대한민국)'])))
y = df['GO: (대한민국)']
y1 = df['python: (대한민국)']
y2 = df['C: (대한민국)']

plt.plot(x, y, color='blue', label='GO')
plt.plot(x1, y1, color='red', label='Python')
plt.plot(x2, y2, color='green', label='C')
plt.savefig('plot.png')
plt.show()
```

```python
# 막데그래프
plt.bar(df['python: (대한민국)'], df['python: (대한민국)'])
plt.title('python')
plt.xlabel('python')
plt.ylabel('python')
```

```python
3D Scatter Chart#히스토그램
# 히스토그램 생성
plt.hist(df['python: (대한민국)'], bins=5)

# 그래프 제목 및 축 제목 설정
plt.title('python:')
plt.xlabel('python:')
plt.ylabel('python user')

# 그래프 표시
plt.show()
```
