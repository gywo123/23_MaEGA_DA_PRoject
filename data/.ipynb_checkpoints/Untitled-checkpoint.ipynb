```python
#3D Scatter Chart
# Fixing random state for reproducibility
np.random.seed(19680801)


def randrange(n, vmin, vmax):
    """
    Helper function to make an array of random numbers having shape (n, )
    with each number distributed Uniform(vmin, vmax).
    """
    return (vmax - vmin)*np.random.rand(n) + vmin

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

n = 100

# For each set of style and range settings, plot n random points in the box
# defined by x in [23, 32], y in [0, 100], z in [zlow, zhigh].
for m, zlow, zhigh in [('o', -50, -25), ('^', -30, -5)]:
    xs = randrange(n, 23, 32)
    ys = randrange(n, 0, 100)
    zs = randrange(n, zlow, zhigh)
    ax.scatter(xs, ys, zs, marker=m)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plts.show()
```

```python
from matplotlib import cm
import numpy as np

from mpl_toolkits.mplot3d.axes3d import get_test_data


# set up a figure twice as wide as it is tall
fig = plt.figure(figsize=plt.figaspect(0.5))

# =============
# First subplot
# =============
# set up the axes for the first plot
ax = fig.add_subplot(1, 2, 1, projection='3d')

# plot a 3D surface like in the example mplot3d/surface3d_demo
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.set_zlim(-1.01, 1.01)
fig.colorbar(surf, shrink=0.5, aspect=10)

# ==============
# Second subplot
# ==============
# set up the axes for the second plot
ax = fig.add_subplot(1, 2, 2, projection='3d')

# plot a 3D wireframe like in the example mplot3d/wire3d_demo
X, Y, Z = get_test_data(0.05)
ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)

plts.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib import animation, rc
```

```python
fig, ax = plt.subplots()

ax.set_xlim(( 0, 2))
ax.set_ylim((-2, 2))

line, = ax.plot([], [], lw=2)
```

```python
def init():
    line.set_data([], [])
    return (line,)
```

```python
def animate(i):
    x = np.linspace(0, 2, 1000)
    y = np.sin(2 * np.pi * (x - 0.01 * i))
    line.set_data(x, y)
    return (line,)
```

```python
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=100, interval=20, blit=True)
```

```python
# Generate heatmap data
rows = 40
columns = 60

Value = namedtuple('Value', ['x', 'z', 'y'])
values = np.random.randint(0, 150, size=(rows, columns))
data = []

for row, column in product(range(rows), range(columns)):
    value = values[row][column]
    points_amount = int(value / 100)
    y_min = 0
    y_max = value
    for _ in range(points_amount):
        y = random.uniform(y_min, y_max)
        data.append(Value(row, column, y))
    data.append(Value(row, column, y_max))

# Convert data to BufferGeometry
geometry = p3.BufferGeometry(
    attributes={
        'position': p3.BufferAttribute(
            np.array([(v.x, v.y, v.z) for v in data], dtype=np.float32),
            normalized=False
        )
    }
)

# Create Points material
material = p3.PointsMaterial(
    size=0.5,
    vertexColors='VertexColors',
    sizeAttenuation=False
)

# Create Points object
points = p3.Points(geometry=geometry, material=material)

# Create pythreejs scene
scene = p3.Scene()
scene.add(points)

# Create pythreejs camera
camera = p3.PerspectiveCamera(position=[10, 10, 10], aspect=1)

# Create pythreejs controls
controls = p3.OrbitControls(controlling=camera)

# Create pythreejs renderer
renderer = p3.Renderer(scene=scene, camera=camera, controls=[controls])

# Display the renderer
display(renderer)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation

ds = pd.read_csv("Automobile.csv")

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(ds['horsepower'], ds['weight'], ds['acceleration'])

ax.set_xlabel('horsepower')
ax.set_ylabel('weight')
ax.set_zlabel('acceleration')

def update_rotation(num, scatter, ax):
    ax.view_init(elev=num, azim=num)  # 변경된 시점을 설정합니다.
    return scatter,

animation = FuncAnimation(fig, update_rotation, frames=range(0, 360, 2),
                          fargs=(scatter, ax), interval=50)

plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from ipywidgets import interactive

ds = pd.read_csv("Automobile.csv")

def create_3d_scatter_plot(angle):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    ax.scatter(ds['horsepower'], ds['weight'], ds['acceleration'])

    ax.set_xlabel('horsepower')
    ax.set_ylabel('weight')
    ax.set_zlabel('acceleration')
    ax.view_init(azim=angle)  # 회전 각도 설정

    plt.show()

# 상호작용 그래프 생성
interactive_plot = interactive(create_3d_scatter_plot, angle=(0, 360, 5))
output = interactive_plot.children[-1]
output.layout.height = '350px'
interactive_plot
```
